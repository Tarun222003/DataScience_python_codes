# Task 2: Dimensionality Reduction with PCA:

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
```

```python
df_new = pd.DataFrame(SS.fit_transform(df))
df_new.columns = list(df)
df = df_new
df
```

```python
from sklearn.decomposition import PCA
pca = PCA()
```

```python
pc = pca.fit_transform(df)
pc = pd.DataFrame(pc)
pc
```

```python
#2.Determine the optimal number of principal components using techniques like scree plot or cumulative explained variance.
import numpy as np
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 6))
plt.plot(range(1, len(explained_variance)+1), explained_variance, marker='o')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.show()
```

# Task 3: Clustering with Original Data:

```python
X = df.iloc[:,1:]
```

```python
from sklearn.cluster import KMeans
K_cluster = KMeans(n_clusters=3)
```

```python
df["K_3"] = K_cluster.fit_predict(X)
```

```python
plt.scatter(df['Alcohol'], df['Proline'], c=df["K_3"], cmap='rainbow')
plt.xlabel('Alcohol')
plt.ylabel('Proline')
plt.show()
```

```python
import numpy as np
from sklearn.metrics import silhouette_score
score=silhouette_score(X,df["K_3"])
print("K_3 score:", np.round(score,3))
```

```python
scores = []
```

```python
for i in range(2,11,1):
    from sklearn.cluster import KMeans
    K_cluster = KMeans(n_clusters=i)
    df["K_3"] = K_cluster.fit_predict(X)
    from sklearn.metrics import silhouette_score
    scores.append(silhouette_score(X,df["K_3"]))
```

```python
np.round(scores,3)
```

```python
import matplotlib.pyplot as plt
plt.scatter(range(2,11),scores)    
plt.plot(range(2,11),scores,color='red')
plt.show()
```
