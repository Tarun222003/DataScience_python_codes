# 2. Encoding Techniques:

```python
OHE_cols = pd.concat([df["sex"],df["income"]],axis=1)
OHE_cols
```

```python
from sklearn.preprocessing import OneHotEncoder
OHE = OneHotEncoder()
```

```python
OHE1 = OHE.fit_transform(OHE_cols[["sex"]]).toarray()
OHE1 = pd.DataFrame(OHE1)
OHE1.columns = ["Sex_Female",'Sex_Male']
OHE1.head()
```

```python
OHE2 = OHE.fit_transform(OHE_cols[["income"]]).toarray()
OHE2 = pd.DataFrame(OHE2)
OHE2.columns = ["income_<=50K",'income_>50K']
OHE2.head()
```

```python
OHE = pd.concat([OHE1,OHE2],axis=1)
OHE
```

```python
objects = df.select_dtypes(include=["object"])
LE_obj = objects.drop(["sex","income"],axis=1)
LE_obj
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
LE_var = LE_obj.apply(LE.fit_transform)
LE_var.columns = list(LE_obj)
LE_var
```

```python
Cat_var = pd.concat([OHE,LE_var],axis=1)
Cat_var
```

```python
df_new = pd.concat([Cat_var,MM_num],axis=1)
df_new
```

# 4. Feature Selection:

```python
from sklearn.ensemble import IsolationForest
iso_forest = IsolationForest(contamination='auto', random_state=42)
```

```python
outliers = iso_forest.fit_predict(df_new[['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']])
```

```python
#Outliers leads to overfitting, skewing, and less model accuracy.
```

```python
#removing outliers
df_no_outliers = df_new[outliers == 1].copy()
```

```python
df_new.shape
```

```python
df_no_outliers.shape
```

```python
df_no_outliers
```

```python
Y = df_no_outliers["income_>50K"]
X = df_no_outliers.drop(["income_>50K","income_<=50K"], axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier   # decision tree regressor
model = DecisionTreeClassifier(criterion='gini') # entropy
model.fit(X,Y)
```

```python
Y_pred =  model.predict(X)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(Y, Y_pred)
mae_baseline = mean_absolute_error(Y_pred, np.full_like(Y_pred, Y.mean()))
```

```python
pps_score = max(0, 1 - (mae / mae_baseline))
print("pps score: ",pps_score)
```
