```python
df["Lifespan (Hours)"].describe()
```

```python
from statsmodels.stats.weightstats import ztest
```

```python
population_mean = 1000
alpha = 0.05  # level of significance
```

```python
#step 1: Perform one sample ztest
z_stat, p_value = ztest(df["Lifespan (Hours)"], value=population_mean, alternative="smaller")
print(z_stat)
print(p_value)
```

```python
if(p_value<alpha):
    print("Ho is rejected ")
else:
    print("H1 is accepted")
```
