```python
import pandas as pd
df = pd.read_csv(r"C:\Users\padal\OneDrive\Desktop\Data Science\Lungcapdata.csv")
df
```

```python
mean = df["LungCap"].mean()
std = df["LungCap"].std()
print(mean)
print(std)
```

```python
from scipy.stats import norm
CI = norm.interval(confidence=0.95, loc=mean, scale=std)
print(f"We are 95% confident that lungcap mean will lie under {CI} ")
```
