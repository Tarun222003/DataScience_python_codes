```python
import pandas as pd
df = pd.read_csv(r"C:\Users\padal\OneDrive\Desktop\Data Science\breast_cancer.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
```

```python
df["Class"] = LE.fit_transform(df["Class"])
Y = df["Class"]
Y
```

```python
X = df.iloc[:,1:10]
X
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
SS_X = SS.fit_transform(X)
SS_X = pd.DataFrame(SS_X)
SS_X.columns = list(X)
SS_X
```

```python
df1 = pd.concat([SS_X,Y],axis=1)
df1
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(SS_X,Y)
```

```python
Y_Pred= model.predict(SS_X)
df["Y_Pred"]
```

```python
Pred_Prob = model.predict_proba(SS_X)[:,1:]
df["Pred_Prob"]
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(Y,Y_Pred)
cm
```

```python
import numpy as np
score = accuracy_score(Y,Y_Pred)
print("Accuracy score:",np.round(score,3))
```

```python
from sklearn.metrics import confusion_matrix,recall_score
r_score = recall_score(Y,Y_Pred)
print("Recall score:",np.round(r_score,3))
```

```python
TN = cm[0,0]
FP = cm[1,0]
TNR = TN/(TN+FP)
print("Specificity:",np.round(TNR,3))
```

```python
from sklearn.metrics import f1_score,precision_score
f1_score = f1_score(Y,Y_Pred)
precision_score = precision_score(Y,Y_Pred)
print("f1 score:",np.round(f1_score,3))
print("precision score:",np.round(precision_score,3))
```

```python
from sklearn.metrics import roc_curve,roc_auc_score
TPR,FPR,dummy = roc_curve(Y,df["Pred_Prob"])
TPR,FPR,dummy
```

```python
import matplotlib.pyplot as plt
plt.scatter(TPR,FPR)
plt.plot(TPR,FPR,color="black")
plt.xlabel("True Positive Rate")
plt.ylabel("True Negitive Rate")
plt.show()
```

```python
auc_score = roc_auc_score(Y,df["Pred_Prob"])
print("AUC Curve:",np.round(auc_score,3))
```
