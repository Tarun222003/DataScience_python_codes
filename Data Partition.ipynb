```python
import pandas as pd
df = pd.read_csv(r"C:\Users\padal\OneDrive\Desktop\Data Science\breast_cancer.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
```

```python
df["Class"] = LE.fit_transform(df["Class"])
Y = df["Class"]
Y
```

```python
X = df.iloc[:,1:10]
X
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
SS_X = SS.fit_transform(X)
SS_X = pd.DataFrame(SS_X)
SS_X.columns = list(X)
SS_X
```

```python
#Data Partition
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(SS_X,Y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
Y_pred_train = model.predict(X_train)
Y_pred_test = model.predict(X_test)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score
score1 = accuracy_score(Y_train,Y_pred_train)
print("Training Accuracy score:",np.round(score1,3))
score2 = accuracy_score(Y_test,Y_pred_test)
print("Testing Accuracy score:",np.round(score2,3))
```

```python
#Cross validation ---->Shuffle split
train_acc = []
test_acc = []
for i in range(1,101):
    X_train,X_test,Y_train,Y_test = train_test_split(SS_X,Y,test_size=0.3,random_state=i)
    model.fit(X_train,Y_train)
    Y_pred_train = model.predict(X_train)
    Y_pred_test = model.predict(X_test)
    train_acc.append(accuracy_score(Y_train,Y_pred_train))
    test_acc.append(accuracy_score(Y_test,Y_pred_test))
```

```python
print("Shfulle split cross validation traing accuracy:",np.round(np.mean(train_acc),2))
print("Shfulle split cross validation testing accuracy:",np.round(np.mean(test_acc),2))
```

```python
#K-Fold Cross Validation
from sklearn.model_selection import KFold
kf = KFold(n_splits = 5) 
```

```python
training_acc = []
testing_acc = []

for train_index, test_index in kf.split(SS_X):
        X_train,X_test,Y_train,Y_test = SS_X.iloc[train_index],SS_X.iloc[test_index],Y[train_index],Y[test_index] 
        model.fit(X_train,Y_train)
        Y_pred_train = model.predict(X_train)
        Y_pred_test = model.predict(X_test)
        training_acc.append(accuracy_score(Y_train,Y_pred_train))
        testing_acc.append(accuracy_score(Y_test,Y_pred_test))
```

```python
print("K-Fold cross validation traing accuracy:",np.round(np.mean(training_acc),2))
print("K-Fold cross validation testing accuracy:",np.round(np.mean(testing_acc),2))
```

```python
from sklearn.metrics import log_loss
kf = KFold(n_splits = 5) 
```

```python
training_acc = []
testing_acc = []
```

```python
for train_index, test_index in kf.split(SS_X):
        X_train,X_test,Y_train,Y_test = SS_X.iloc[train_index],SS_X.iloc[test_index],Y[train_index],Y[test_index] 
        model.fit(X_train,Y_train)
        Y_pred_train = model.predict(X_train)
        Y_pred_test = model.predict(X_test)
        training_acc.append(log_loss(Y_train,Y_pred_train))
        testing_acc.append(log_loss(Y_test,Y_pred_test))
```

```python
print("Log loss for traing accuracy:",np.round(np.mean(training_acc),2))
print("Log loss for testing accuracy:",np.round(np.mean(testing_acc),2))
```
