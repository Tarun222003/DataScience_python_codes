```python
import pandas as pd
import numpy as np
df = pd.read_csv(r"C:\Users\padal\OneDrive\Desktop\Data Science\mushroom.csv")
df
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
```

```python
df = df.apply(LE.fit_transform)
df
```

```python
X = df.iloc[:,1:23]
Y = df["Typeofmushroom"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
```

```python
model.fit(X_train,Y_train)
```

```python
Y_pred_train = model.predict(X_train)
Y_pred_test = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
training_acc = accuracy_score(Y_train,Y_pred_train)
testing_acc = accuracy_score(Y_test,Y_pred_test)
print("Training accuracy:",np.round(training_acc,2))
print("Testing accuracy:",np.round(testing_acc,2))
```

```python
train_acc = []
test_acc = []

for i in range(1,201):
    X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3)
    model = MultinomialNB()
    model.fit(X_train,Y_train)
    Y_pred_train = model.predict(X_train)
    Y_pred_test = model.predict(X_test)
    training_acc = accuracy_score(Y_train,Y_pred_train)
    testing_acc = accuracy_score(Y_test,Y_pred_test)
    train_acc.append(training_acc)
    test_acc.append(testing_acc)
```

```python
print("shuffle split cross validation for training data",np.round(np.mean(train_acc),2))
print("shuffle split cross validation for testing data",np.round(np.mean(test_acc),2))
```
