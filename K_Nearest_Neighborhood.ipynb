```python
import pandas as pd
df = pd.read_csv(r"C:\Users\padal\OneDrive\Desktop\Data Science\breast-cancer-wisconsin-data.csv")
df
```

```python
df.info()
```

```python
X = df.iloc[:,2:32]
X
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
SS_X = SS.fit_transform(X)
SS_X = pd.DataFrame(SS_X)
SS_X.columns = list(X)
SS_X
```

```python
Y = df["diagnosis"]
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
LE_Y = LE.fit_transform(Y)
LE_Y = pd.DataFrame(LE_Y)
LE_Y.columns = ["diagnosis"]
LE_Y
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pc = pca.fit_transform(SS_X)
pc = pd.DataFrame(pc)
pc
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import numpy as np
model = LogisticRegression()
```

```python
train_acc = []
test_acc = []
```

```python
for i in range(1,101):
    X_train,X_test,Y_train,Y_test =train_test_split(SS_X,Y,test_size=0.30,random_state=i)
    model.fit(X_train,Y_train)
    Y_pred_train =  model.predict(X_train)
    Y_pred_test =  model.predict(X_test)
    train_acc.append(accuracy_score(Y_train,Y_pred_train))
    test_acc.append(accuracy_score(Y_test,Y_pred_test))
```

```python
print("shuffle split-cv-training accuracy:",np.round(np.mean(train_acc),2))
print("shuffle split-cv-test accuracy:",np.round(np.mean(test_acc),2))
```

```python
training_accuracy_list = []
testing_accuracy_list = []
```

```python
for k in range(5,18,2):
    
    training_accuracy = []
    testing_accuracy = []
    
    for i in range(1,101):
        
        X_train,X_test,Y_train,Y_test =train_test_split(X_new,Y,test_size=0.30,random_state=i)
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train,Y_train)
        Y_pred_train =  knn.predict(X_train)
        Y_pred_test =  knn.predict(X_test)
        training_accuracy.append(accuracy_score(Y_train,Y_pred_train))
        testing_accuracy.append(accuracy_score(Y_test,Y_pred_test))
    
training_accuracy_list.append(np.mean(training_accuracy).round(2))
testing_accuracy_list.append(np.mean(testing_accuracy).round(2))
```
