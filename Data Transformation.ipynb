```python
df.info()
```

```python
X_cont = df.iloc[:,2:6]
X_cont.head()
```

```python
# standardization
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
SS_X = SS.fit_transform(X_cont)
SS_X = pd.DataFrame(SS_X)
SS_X.columns = list(X_cont)
SS_X.head()
```

```python
# scaling
from sklearn.preprocessing import MinMaxScaler
MM = MinMaxScaler()
MM_X = MM.fit_transform(X_cont)
MM_X = pd.DataFrame(MM_X)
MM_X.columns = list(X_cont)
MM_X.head()
```

```python
# label encoding
X_cat = df[["Gender","Diabetic"]]
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
X_cat["Gender"] = LE.fit_transform(X_cat["Gender"])
X_cat["Diabetic"] = LE.fit_transform(X_cat["Diabetic"])
X_cat.head()
```

```python
# one hot encoding
X_cat2 = df[["Gender","Diabetic"]]
X_cat2
```

```python
from sklearn.preprocessing import OneHotEncoder
OHE = OneHotEncoder()
```

```python
OH2 = OHE.fit_transform(X_cat2[["Gender"]]).toarray()
OH2 = pd.DataFrame(OH2)
OH2.columns = ["G_Female",'G_Male']
OH2.head()
```

```python
OH3 = OHE.fit_transform(X_cat2[["Diabetic"]]).toarray()
OH3 = pd.DataFrame(OH3)
OH3.columns = ["D_No",'D_Yes']
OH3.head()
```

```python
df_new = pd.concat([SS_X,X_cat],axis=1)
df_new
```
