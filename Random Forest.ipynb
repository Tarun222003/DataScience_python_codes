```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
SS_X = SS.fit_transform(X)
SS_X = pd.DataFrame(SS_X)
SS_X.columns = list(X)
SS_X
```

```python
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='accuracy')
```

```python
grid_search.fit(X_train, Y_train)
```

```python
print("Best parameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
import numpy as np
```

```python
train_acc = []
test_acc = []
```

```python
for i in range(1,101):
    from sklearn.ensemble import RandomForestClassifier
    model = RandomForestClassifier(n_estimators=100, random_state=42,max_samples=0.7,max_features=0.6)
    model.fit(X_train,Y_train)
    Y_pred_train = model.predict(X_train)
    Y_pred_test = model.predict(X_test)
    train_acc.append(accuracy_score(Y_train,Y_pred_train))
    test_acc.append(accuracy_score(Y_test,Y_pred_test))
```

# Additional Notes:
1. Explain Bagging and Boosting methods. How is it different from each other.
Ans)Bagging: Train multiple models in parallel on different random subsets of the training data (sampled with replacement) and then combine their            predictions.
    Boosting: Train models sequentially. Each new model focuses more on the errors (misclassified points) made by the previous ones.

   Bagging = "Divide and Conquer" â†’ Many models in parallel, then combine.
    Boosting = "Learn from mistakes" â†’ Train one after another, each correcting the previous.
